```python
moyenne_prix = sum(maison[1] for maison in prix_surface) / len(prix_surface)
moyenne_surface = sum(maison[0] for maison in prix_surface) / len(prix_surface)
covariance = sum((maison[1] - moyenne_prix) * (maison[0] - moyenne_surface) for maison in prix_surface)/len(prix_surface)

# Calcul des ecarts_types
et_prix = (sum((maison[1] - moyenne_prix) ** 2 for maison in prix_surface)/len(prix_surface))**0.5
et_surface = (sum((maison[0] - moyenne_surface) ** 2 for maison in prix_surface)/len(prix_surface))**0.5

# Calcul du coefficient de corrélation
coefficient_correlation = covariance/(et_prix*et_surface)

print("Le coefficient de corrélation entre le prix et la surface des maisons est :", coefficient_correlation)
```
